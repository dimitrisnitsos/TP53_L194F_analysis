```python
# Calculate the complementary DNA sequence
def complement(Nucleotide):
    comp = []
    for i in Nucleotide:
        if i == "T":
            comp.append("A")
        if i == "A":
            comp.append("T")
        if i == "G":
            comp.append("C")
        if i == "C":
            comp.append("G")
        if i == "t":
            comp.append("a")
        if i == "a":
            comp.append("t")
        if i == "g":
            comp.append("c")
        if i == "c":
            comp.append("g")    
    return ''.join(comp)
```

```python
#find the NGG nucleotide indexes

def list_of_canditated_PAM(string, substring):     # string could be coding strand or non coding strand, here we examine only coding strand
  
    # Initialize count and start to 0
    count = 0
    start = 0
    lista = []
  
    # Search through the string till
    # we reach the end of it
    while start < len(string):
  
        # Check if a substring is present from
        # 'start' position till the end
        pos = string.find(substring, start)
  
        if pos != -1:
            
            # If a substring is present, move 'start' to
            # the next position from start of the substring
            start = pos + 1
            if substring == 'GG':
                lista.append(start-2)
            if substring == 'CC':
                lista.append(start-1)
            # Increment the count
            count += 1
        else:
            # If no further substring is present
            break
    # return the value of count
    return lista
```

```python
# Μultiplication of all possible RT templates X all possible PBS in order to create the 3' extension


def dict_indexPAM_calculate_RT_PBS(dictionary_indexPAM_PAM_PBS, dictionary_indexPAM_PAM_RT):

    dicts = {}
    i = k =0
    temp_calculations_multiplication = []
    for key in dictionary_indexPAM_PAM_PBS.keys():
        #print("key = ", key)
                
        for i in range(0, len(dictionary_indexPAM_PAM_PBS.get(key)),1):
            #print("value = ",dictionary_indexPAM_PAM_PBS.get(key)[i])
            value_PBS = dictionary_indexPAM_PAM_PBS.get(key)[i]
            # print("\n")
            for k in range(0, len(dictionary_indexPAM_PAM_RT.get(key)),1):
                #print("value = ",dictionary_indexPAM_PAM_RT.get(key)[k])
                value_RT = dictionary_indexPAM_PAM_RT.get(key)[k]
                add = value_RT + value_PBS
                #print("add = ", add)
                temp_calculations_multiplication.append(add)
                #print("temp_calculations_multiplication = ", temp_calculations_multiplication)
                dicts[key] = temp_calculations_multiplication
            
    return dicts

dictionary_indexPAM_NGG_RT_PBS = dict_indexPAM_calculate_RT_PBS(dictionary_indexPAM_NGG_PBS, dictionary_indexPAM_NGG_RT)
dictionary_indexPAM_CCN_RT_PBS = dict_indexPAM_calculate_RT_PBS(dictionary_indexPAM_CCN_PBS, dictionary_indexPAM_CCN_RT)
```
